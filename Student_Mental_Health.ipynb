```python
import kagglehub
import os
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("shariful07/student-mental-health")
csv_file=os.path.join(path,'Student Mental health.csv')
df=pd.read_csv(csv_file)
print("Location :",os.listdir(path))
# print("Path to dataset files:", path)
```

```python
# df['Choose your gender'].unique()
value_count = df['What is your course?'].value_counts()

def simplify_course(course):
    count = value_count.get(course, 0)
    if count == 18:
        return "BCS"
    elif count == 17:
        return "Engineering"
    elif count == 10:
        return "BIT"
    else:
        return "Others"

df['What is your course?'] = df['What is your course?'].apply(simplify_course)
```

```python
df['What is your course?'].value_counts()
```

```python
df['Your current year of Study'] = df['Your current year of Study'].astype(str).str.strip().str.lower()
```

```python
df['What is your CGPA?'] = df['What is your CGPA?'].str.strip()

df['What is your CGPA?'] = df['What is your CGPA?'].apply(
    lambda x: '3.50 - 4.00' if x == '3.50 - 4.00' else x
)

# def simply(CGPA):
#   count = value_count.get(CGPA, 0)
#   if '3.50 - 4.00' or '3.50 - 4.00'
```

```python
from sklearn.model_selection import train_test_split
# from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import confusion_matrix,accuracy_score
x,y=df.drop('Did you seek any specialist for a treatment?',axis=1),df['Did you seek any specialist for a treatment?']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
lg=DecisionTreeClassifier(criterion='entropy')
lg.fit(x_train,y_train)
y_pred=lg.predict(x_test)
acc=(accuracy_score(y_test,y_pred))
acc
```
